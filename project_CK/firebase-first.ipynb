```python
cred = credentials.Certificate("./iuh-20089981-firebase-adminsdk-7x1ba-005c4884a4.json")
app = firebase_admin.initialize_app(cred)
```

```python
dbFireStore = firestore.client()
```

```python
# load file data lên database 
for item in df.to_dict(orient='records'):
    doc_ref = dbFireStore.collection("tblIUHSALES").document()
    doc_ref.set(item)
```

```python
#cau du lieu trong NoSQL

queryResults = list(dbFireStore.collection(u'tblIUHSALES').limit(5).stream())
listQueryResult = list(map(lambda x : x.to_dict(), queryResults))
df = pd.DataFrame(listQueryResult)
df
```

```python
yResults = list(dbFireStore.collection(u'tblIUHSALES').where(u'YEAR_ID', u'>=', 2005).limit(10).stream())
QueryResult = list(map(lambda x : x.to_dict(), yResults))
df = pd.DataFrame(QueryResult)
df
```

```python
where2Results = list(dbFireStore.collection(u'tblIUHSALES').where(u'YEAR_ID', u'==', 2005).where(u'COUNTRY', u'==', "USA").limit(10).stream())
QueryResultWhere = list(map(lambda x : x.to_dict(), where2Results))
df = pd.DataFrame(QueryResultWhere)
df
```
